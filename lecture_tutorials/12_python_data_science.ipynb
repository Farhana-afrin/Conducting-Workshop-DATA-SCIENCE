## contatenation

```python
import pandas as pd
```

```python
BD_age_height = pd.DataFrame({
    "classes": ["class1", "class3", "class5", "class7", "class9", "class12", "undergrad_done"],
    "age": [7.27, 9.31, 11.79, 13.76, 16.01, 19.95, 27.15],
    "height": [49.23, 54.01, 56.90, 58.77 ,59.13 ,60.91, 62.05]
})
BD_age_height
```

```python
US_age_height = pd.DataFrame({
    "classes": ["class1", "class3", "class5", "class7", "class9", "class12", "undergrad_done"],
    "age": [6.77, 8.71, 10.79, 12.76, 15.01, 18.75, 24.15],
    "height": [49.93, 55.01, 57.77, 60.97 ,61.13 ,63.01, 63.85]
})
US_age_height
```

```python
China_age_height = pd.DataFrame({
    "classes": ["class1", "class3", "class5", "class7", "class9", "class12", "undergrad_done"],
    "age": [5.77, 7.71, 9.79, 11.76, 14.01, 17.75, 22.15],
    "height": [48.93, 54.01, 56.77, 58.97 ,59.13 ,61.01, 62.85]
})
China_age_height
```

```python
df = pd.concat([BD_age_height, US_age_height, China_age_height], axis=0)
df
```

```python
df = pd.concat([BD_age_height, US_age_height, China_age_height], axis=0, ignore_index=True)
df
```

```python
df = pd.concat([BD_age_height, US_age_height, China_age_height], axis=1)
df
```

```python
df_concat = pd.concat([BD_age_height, US_age_height, China_age_height], axis=0, keys=['Bangladesh', 'USA', 'China'])
df_concat
```

```python
df_concat.loc['China']
```

```python
# Adding a new column:
col_gender = pd.Series(["Female","Female","Female", "Female", "Female", "Male", "Male"], name="max_gender")
pd.concat([BD_age_height, col_gender], axis=1)
```

```python
pd.merge(BD_age_height, US_age_height, on="classes", suffixes=('_BD', '_US'))
```

```python
# Let's create some dataframes

Book_and_Author = pd.DataFrame({'Authors':['H Ahmed', 'H Ahmed', 'H Ahmed', 'MZI', 'R Tagore', 'R Tagore', 'S chattopadhyay', 'R Tagore'],
                  'Books':['1971', 'Tomake', 'Deyal', 'Tukunjil', 'Gitanjali', 'Chitrangoda', 'Devdas','Chelebela']})
Book_and_Author
```

```python
Book_and_Published = pd.DataFrame({'Books':['Gabhi Bittanta', 'Nabani', 'Deyal', 'Tukunjil', 'Gitanjali', 'Chitrangoda', 'Devdas','Chelebela','Parinita'],
                     'Published':[1994, 1993, 2013, 2000, 1910, 1892, 1917, 1940, 1916]})
Book_and_Published
```

```python
#inner join
df_inner = pd.merge(Book_and_Author, Book_and_Published, on="Books", how="inner")
df_inner
```

```python
#Outer join
df_outer = pd.merge(Book_and_Author, Book_and_Published, on="Books", how="outer")
df_outer
```

```python
#left join
df_left = pd.merge(Book_and_Author, Book_and_Published, on="Books", how="left")
df_left
```

```python
#right join
df_right = pd.merge(Book_and_Author, Book_and_Published, on="Books", how="right")
df_right
```

```python
#right join
df_right = pd.merge(Book_and_Author, Book_and_Published, on="Books", how="right", indicator=True)
df_right
```

```python
Book_and_Published.set_index('Books').join(Book_and_Author.set_index('Books'))
```
