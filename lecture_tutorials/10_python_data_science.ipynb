## How to generate pandas dataframe

```python
#Way 1: From the list of lists

## Initialize your data
data = [['Andrew', 'USA', 44], ['Prity', 'India', 32], ['Arijit', 'India', 28], ['Mohammad', 'Egypt', 37], ['Imara', 'Srilanka', 29]]

## Initialize the dataframe
df = pd.DataFrame(data, columns = ['Name', 'Counry', 'Age'])
df
```

```python
# Way2: Throught the dictionary
data = {'Name':['Andrew', 'Prity', 'Arijit', 'Mohammad', 'Imara'],
       'Country':['USA', 'India', 'India', 'Egypt', 'Imara'],
       'Age':[44, 32, 28, 37, 29]}
df = pd.DataFrame(data)
df
```

```python
# You can specify the indexes if you want
df = pd.DataFrame(data, index =['student1', 'student2', 'student3', 'student4', 'student5'])
df
```

```python
#Way 3: List of dicts
data = [{'NUM': 1, 'SQRT': 1, 'CUBERT':1}, 
        {'NUM': 64, 'SQRT': 8, 'CUBERT':4}, 
        {'NUM': 729, 'SQRT': 27, 'CUBERT':9}, 
        {'NUM': 4096, 'SQRT': 64, 'CUBERT':16}] 
df = pd.DataFrame(data)
df
```

```python
#Way 4: tuples
# List1  
Name = ['Andrew', 'Prity', 'Arijit', 'Mohammad', 'Imara']
Country = ['USA', 'India', 'India', 'Egypt', 'Imara']
Age = [44, 32, 28, 37, 29]

    
# get the list of tuples from two lists.    
list_of_tuples = list(zip(Name, Age, Country))  
df = pd.DataFrame(list_of_tuples, columns = ['Name', 'Age', 'Country'])  
     
# Print data.  
df  
```

```python
# way 5: Pandas Series
data = {'day':pd.Series(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday']),
    'temperature':pd.Series([20, 28, 21, 22, 29])}
df = pd.DataFrame(data)
df
```

```python
df = df.set_index('day')
df
```
